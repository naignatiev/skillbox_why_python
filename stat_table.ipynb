# Генерация датасета

```python
from names import get_full_name
from random import randint
from pandas import DataFrame, read_csv


def get_phone_number():
    return '+79' + ''.join([str(randint(0, 9)) for i in range(9)])
    

ROWS_COUNT = 200

data = []
for i in range(ROWS_COUNT):
    full_name = get_full_name()
    phone_number = get_phone_number()
    monday = randint(0, 1)
    tuesday = randint(0, 1)
    wednesday = randint(0, 1)
    data.append((full_name, phone_number, monday, tuesday, wednesday))


table = DataFrame(data, columns=['full_name', 'phone_number', 'monday', 'tuesday', 'wednesday'])
display(table)
```

```python
table.to_csv('table.csv')
```
